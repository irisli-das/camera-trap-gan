```python
import cv2
import os

def extract_frames(video_path, output_folder):
    """
    Extract ALL frames from a video file.
    """
    os.makedirs(output_folder, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_fps = cap.get(cv2.CAP_PROP_FPS)

    print(f"Video: {os.path.basename(video_path)}")
    print(f"  FPS: {video_fps} | Total frames: {total_frames}")

    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        filename = os.path.join(output_folder, f"frame_{frame_count:05d}.jpg")
        cv2.imwrite(filename, frame)
        frame_count += 1

        # progress update every 1000 frames
        if frame_count % 1000 == 0:
            print(f"  saved {frame_count}/{total_frames} frames...")

    cap.release()
    print(f"  done — extracted {frame_count} frames to {output_folder}\n")
    return frame_count


def extract_all_videos(video_folder, output_folder):
    """
    Extract ALL frames from every video in a folder.
    Each video gets its own subfolder.
    """
    video_extensions = ['.mp4', '.avi', '.mov', '.mkv', '.MP4', '.AVI', '.MOV']

    # find all videos
    videos = [f for f in os.listdir(video_folder)
              if any(f.endswith(ext) for ext in video_extensions)]

    if not videos:
        print(f"No videos found in {video_folder}")
        return

    print(f"Found {len(videos)} videos\n")
    total_frames = 0

    for i, filename in enumerate(sorted(videos)):
        print(f"[{i+1}/{len(videos)}] Processing {filename}")
        video_path   = os.path.join(video_folder, filename)
        video_name   = os.path.splitext(filename)[0]
        video_output = os.path.join(output_folder, video_name)

        frames = extract_frames(video_path, video_output)
        total_frames += frames

    print(f"Finished! Total frames extracted: {total_frames:,}")
    print(f"Saved to: {output_folder}")


if __name__ == "__main__":
    extract_all_videos(
        video_folder="/Users/iris/Desktop/ETH/DAS Project/camera-trap-gan/das_birds/videos",
        output_folder="/Users/iris/Desktop/ETH/DAS Project/camera-trap-gan/data/frames"
    )
```
