```python
import cv2
import os

def extract_frames(video_path, output_folder):
    """
    Extract ALL frames from a video file.
    """
    os.makedirs(output_folder, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_fps = cap.get(cv2.CAP_PROP_FPS)

    print(f"Video: {os.path.basename(video_path)}")
    print(f"  FPS: {video_fps} | Total frames: {total_frames}")

    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        filename = os.path.join(output_folder, f"frame_{frame_count:05d}.jpg")
        cv2.imwrite(filename, frame)
        frame_count += 1

        # progress update every 1000 frames
        if frame_count % 1000 == 0:
            print(f"  saved {frame_count}/{total_frames} frames...")

    cap.release()
    print(f"  done — extracted {frame_count} frames to {output_folder}\n")
    return frame_count


def extract_all_videos(video_folder, output_folder):
    """
    Extract ALL frames from every video in a folder.
    Each video gets its own subfolder.
    """
    video_extensions = ['.mp4', '.avi', '.mov', '.mkv', '.MP4', '.AVI', '.MOV']

    # find all videos
    videos = [f for f in os.listdir(video_folder)
              if any(f.endswith(ext) for ext in video_extensions)]

    if not videos:
        print(f"No videos found in {video_folder}")
        return

    print(f"Found {len(videos)} videos\n")
    total_frames = 0

    for i, filename in enumerate(sorted(videos)):
        print(f"[{i+1}/{len(videos)}] Processing {filename}")
        video_path   = os.path.join(video_folder, filename)
        video_name   = os.path.splitext(filename)[0]
        video_output = os.path.join(output_folder, video_name)

        frames = extract_frames(video_path, video_output)
        total_frames += frames

    print(f"Finished! Total frames extracted: {total_frames:,}")
    print(f"Saved to: {output_folder}")


if __name__ == "__main__":
    extract_all_videos(
        video_folder="/home/renku/work/videos",
        output_folder= "/home/renku/work/frames"
    )
```

```python
import cv2
import os

# ── specify which videos to process ───────────────────────────
# leave empty [] to process ALL videos in the folder
# or list specific video names to process only those
VIDEOS_TO_PROCESS = [
    "FH706_01",
    "FH803_01"
]
# ──────────────────────────────────────────────────────────────

def extract_frames(video_path, output_folder):
    """
    Extract ALL frames from a video file.
    Skips if output folder already has frames (resume support).
    """
    os.makedirs(output_folder, exist_ok=True)

    # ── resume check ──────────────────────────────────────────
    existing = [f for f in os.listdir(output_folder) if f.endswith('.jpg')]
    if existing:
        print(f"  SKIP: {output_folder} already has {len(existing):,} frames")
        return len(existing)

    cap          = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_fps    = cap.get(cv2.CAP_PROP_FPS)

    print(f"Video: {os.path.basename(video_path)}")
    print(f"  FPS: {video_fps} | Total frames: {total_frames}")

    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        cv2.imwrite(
            os.path.join(output_folder, f"frame_{frame_count:05d}.jpg"), frame)
        frame_count += 1
        if frame_count % 1000 == 0:
            print(f"  saved {frame_count}/{total_frames} frames...")

    cap.release()
    print(f"  done — extracted {frame_count} frames to {output_folder}\n")
    return frame_count


def extract_all_videos(video_folder, output_folder):
    video_extensions = ['.mp4', '.avi', '.mov', '.mkv',
                        '.MP4', '.AVI', '.MOV', '.MKV']

    # find all videos
    all_videos = sorted([
        f for f in os.listdir(video_folder)
        if any(f.endswith(ext) for ext in video_extensions)
    ])

    if not all_videos:
        print(f"No videos found in {video_folder}")
        return

    # filter to specific videos if list is provided
    if VIDEOS_TO_PROCESS:
        videos = [
            f for f in all_videos
            if os.path.splitext(f)[0] in VIDEOS_TO_PROCESS
        ]
        print(f"Found {len(all_videos)} total videos")
        print(f"Processing {len(videos)} selected: {VIDEOS_TO_PROCESS}\n")
        not_found = [v for v in VIDEOS_TO_PROCESS
                     if v not in [os.path.splitext(f)[0] for f in all_videos]]
        if not_found:
            print(f"WARNING: these not found in folder: {not_found}\n")
    else:
        videos = all_videos
        print(f"Processing all {len(videos)} videos\n")

    total_frames = 0
    for i, filename in enumerate(videos):
        print(f"[{i+1}/{len(videos)}] {filename}")
        video_path   = os.path.join(video_folder, filename)
        video_name   = os.path.splitext(filename)[0]
        video_output = os.path.join(output_folder, video_name)
        total_frames += extract_frames(video_path, video_output)

    print(f"Finished! Total frames: {total_frames:,}")
    print(f"Saved to: {output_folder}")


if __name__ == "__main__":
    extract_all_videos(
        video_folder="/home/renku/work/videos",
        output_folder="/home/renku/work/frames"
    )
```
